## Loading NIFTI images using NiBabel

NiBabel is a python library for working with neuro-imaging formats (including NIFTI) that we have used in some of the exercises throughout the course. Our volumes and labels are in NIFTI format, so we will use nibabel to load and inspect them.

NiBabel documentation could be found here: https://nipy.org/nibabel/

Our dataset sits in two directories - *images* and *labels*. Each image is represented by a single file (we are fortunate to have our data converted to NIFTI) and has a corresponding label file which is named the same as the image file.

Note that our dataset is "dirty". There are a few images and labels that are not quite right. They should be quite obvious to notice, though. The dataset contains an equal amount of "correct" volumes and corresponding labels, and you don't need to alter values of any samples in order to get the clean dataset.

```python
# TASK: Your data sits in directory /data/TrainingSet.
def getFileNames(mypath):
    f = []
    for (dirpath, dirnames, filenames) in walk(mypath):        
        filenames = [ f for f in filenames if f.endswith("nii.gz")]        
        f.extend(filenames)
    return f



IMAGES_DIR = "..\data\TrainingSet\images"
LABELS_DIR = "..\data\TrainingSet\labels"

images_list = getFileNames(IMAGES_DIR)
labels_list = getFileNames(LABELS_DIR)
    

# See if there are any missing images/labels
exclusions_list = set(images_list) - set(labels_list)
print(exclusions_list)
```

```python
#remove the extra file that should not be present
for exclusion_file in exclusions_list:
    path = os.path.join(IMAGES_DIR, exclusion_file)
    print(path)
    if os.path.exists(path):
        os.remove(path)
```

```python
def get_file_path(directory,filename):
    path = os.path.join(directory, filename)    
    if os.path.exists(path):
        return path
    return None
```

```python
# Load an image and a segmentation mask into variables called image and label
image = nib.load(get_file_path(IMAGES_DIR,images_list[0])).get_fdata()
label = nib.load(get_file_path(LABELS_DIR,images_list[0])).get_fdata()
image.shape, label.shape
print(images_list[0])
```

```python
#make sure that the shapes of images and labels match and print the list of files that dont match
# reload images as we have cleaned exclusions
images_list = getFileNames(IMAGES_DIR)
labels_list = getFileNames(LABELS_DIR)
mismatched_files=[]
for file in images_list:
    img = nib.load(get_file_path(IMAGES_DIR,file)).get_fdata()
    lbl = nib.load(get_file_path(LABELS_DIR,file)).get_fdata()
    if img.shape != lbl.shape:
        print(file," Image shape:",img.shape," Labels shape:",lbl.shape)
        mismatched_files.append(file)

print(mismatched_files)
```

```python
#delete the mismatched files
for file in mismatched_files:
    img_path = os.path.join(IMAGES_DIR, file)
    print("deleting....",img_path)
    lbl_path = os.path.join(LABELS_DIR, file)      
    if os.path.exists(img_path):
        os.remove(img_path)
    if os.path.exists(lbl_path):
        os.remove(lbl_path)
```

```python
# Nibabel can present your image data as a Numpy array by calling the method get_fdata()
# The array will contain a multi-dimensional Numpy array with numerical values representing voxel intensities. 
# In our case, images and labels are 3-dimensional, so get_fdata will return a 3-dimensional array. You can verify this
# by accessing the .shape attribute. What are the dimensions of the input arrays?
print("shape of the image:",image.shape)
print("shape of the label:",label.shape)


```

```python
# TASK: using matplotlib, visualize a few slices from the dataset, along with their labels. 
# You can adjust plot sizes like so if you find them too small:
# plt.rcParams["figure.figsize"] = (10,10)

# shape of the image is (35,51,35)
def display_volume(image):
    fig, ax = plt.subplots(5, 7, figsize=[20,20])
    img = np.transpose(image, (2, 0, 1))
    for i in range(35):
        ix = i
        ax[int(i/7), int(i%7)].set_title(f"slice {ix}")
        ax[int(i/7), int(i%7)].imshow(img[ix, :, :], cmap='gray')
        ax[int(i/7), int(i%7)].axis("off")

    plt.show()

display_volume(image)
```

## Plotting some charts

```python
# TASK: Plot a histogram of all volumes that we have in our dataset and see how 
# our dataset measures against a slice of a normal population represented by the chart below.
volume = []
for file in labels_list:
    file_path = get_file_path(LABELS_DIR,file)
    if file_path is not None:
        volume.append(np.sum(nib.load(file_path).get_fdata() > 0))
    
plt.hist(volume, bins=50);
plt.title('Histogram of All Volumes');
plt.xlabel('Volume');
plt.show()
```

```python
# You should get a number between ~2200 and ~4500
for val in volume:
    if val<2200 or val > 4500:
        print("Outlier....")
        
#No Outliers were found.
```

```python
# TASK: Copy the clean dataset to the output folder inside section1/out. You will use it in the next Section
def get_full_paths(directory,fileslist):
    full_paths = []
    for file in fileslist:
        file_path = get_file_path(directory,file)
        if file_path is not None:
            full_paths.append(file_path)
    return full_paths
```

```python
image_full_paths = get_full_paths(IMAGES_DIR,images_list)
labels_full_paths = get_full_paths(LABELS_DIR,labels_list)
```

```python
print(len(image_full_paths))
print(len(labels_full_paths))
```

```python
for filename in image_full_paths:
    shutil.copy(filename, "out/images/")
for filename in labels_full_paths:
    shutil.copy(filename, "out/labels/")
```
